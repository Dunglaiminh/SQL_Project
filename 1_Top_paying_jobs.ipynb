```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
# JSON data
data = [
    {"job_title_short": "Senior Data Scientist", "year_salary": "154050"},
    {"job_title_short": "Senior Data Engineer", "year_salary": "145867"},
    {"job_title_short": "Data Scientist", "year_salary": "135929"},
    {"job_title_short": "Data Engineer", "year_salary": "130267"},
    {"job_title_short": "Machine Learning Engineer", "year_salary": "126786"},
    {"job_title_short": "Senior Data Analyst", "year_salary": "114104"},
    {"job_title_short": "Software Engineer", "year_salary": "112778"},
    {"job_title_short": "Cloud Engineer", "year_salary": "111268"},
    {"job_title_short": "Data Analyst", "year_salary": "93876"},
    {"job_title_short": "Business Analyst", "year_salary": "91071"}
]

# Create DataFrame
df = pd.DataFrame(data)

# Convert year_salary to integer
df["year_salary"] = df["year_salary"].astype(int)

# Create a bar plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="year_salary", y="job_title_short", data=df, hue="job_title_short", palette="mako", legend=False, order=df.sort_values("year_salary", ascending=False)["job_title_short"])

# Add labels and title
plt.xlabel("Average Yearly Salary (USD)", fontsize=12)
plt.ylabel("")
plt.title("Top 10 Highest-paid Data-Related Roles", fontsize=14)

# Show exact salary values on bars
for i, v in enumerate(df.sort_values("year_salary", ascending=False)["year_salary"]):
    ax.text(v + 2000, i, f"${v:,}", color='black', va='center', fontsize=10)

# Formatter function
def salary_formatter(x, pos):
    return f'${int(x/1000)}K'
plt.gca().xaxis.set_major_formatter(FuncFormatter(salary_formatter))
ax.xaxis.grid(False)  
ax.yaxis.grid(False) 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show the plot
plt.show()
plt.tight_layout()
```
