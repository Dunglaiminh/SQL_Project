```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
skills = ["SQL", "Excel", "Python", "Tableau", "R", "Power BI", "SAS", "Word", "PowerPoint", "SQL Server"]
demand = [3083, 2143, 1840, 1659, 1073, 1044, 1000, 527, 524, 336]
salaries = [96435, 86419, 101512, 97978, 98708, 92324, 93707, 82941, 88316, 96191]

# Create DataFrame
df = pd.DataFrame({"Skill": skills, "Demand": demand, "Salary": salaries})

# Set style
sns.set_style("whitegrid")

# Plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="Demand", y="Skill", data=df, palette="mako")

# Add salary values at the end of bars
for index, value in enumerate(demand):
    plt.text(value + 50, index, f"${salaries[index]:,}", va="center", fontsize=12, color="black")

# Labels & Title
plt.xlabel("Job Postings Count")
plt.ylabel("")
plt.title("Top 10 In-Demand Data Analyst Skills & Their Salaries")

sns.despine(right=True,top=True)
# Show Plot
plt.show()
plt.tight_layout()
```
