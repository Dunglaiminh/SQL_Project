```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Skills": ["SQL", "Excel", "Python", "Tableau", "Power BI"],
    "Skill Count": [92628, 67031, 57326, 46554, 39468]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set style
sns.set_style("whitegrid")

# Create bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Skill Count", y="Skills", data=df, palette="Blues_r")

# Remove right and top spines
sns.despine(right=True, top=True)
plt.grid(False)

# Add labels
plt.xlabel("Number of Job Postings")
plt.ylabel('')
plt.title("Top 5 Most Common Skills for Data Analysts")

# Show plot
plt.show()
plt.tight_layout()
```
